```python
# Import our dependencies
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
# Import and read data set
application_df = pd.read_csv("Resources/exams.csv")
application_df.head()
```

```python
# Determine the number of unique values in each column.
application_df.nunique()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
dummy_df = pd.get_dummies(application_df)
dummy_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = dummy_df["math score"].values
X = dummy_df.drop(["math score", "reading score", "writing score"], axis=1).values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 1)
```

```python
# Create scaler instance
X_scaler = StandardScaler()

# Fit the scaler
X_scaler.fit(X_train)

# Scale the data via Transform
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```
